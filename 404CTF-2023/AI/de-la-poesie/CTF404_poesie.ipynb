```python
from classifier import ConvNet
import torch
from torchvision import datasets, transforms

model = ConvNet()
model.load_state_dict(torch.load('model/trained.pth'))
print(model)
```

For every image:
1. Load
2. Predict handwritter digit
3. Produce bit value:
  - 0 if even
  - 1 if odd
4. Concatenate bits to form sequence and convert to ASCII

```python
from PIL import Image
import torchvision.transforms.functional as TF

bits = []
for i in range(0, 6536):
  image = Image.open(f'/content/adversarial-mnist/images/{i}.jpg')
  x = TF.to_tensor(image)
  x.unsqueeze_(0)

  output = model(x)
  value = output.argmax()
  b = str(0 if value % 2 == 0 else 1)
  bits.append(b)
```

```python
''.join(bits)
```
